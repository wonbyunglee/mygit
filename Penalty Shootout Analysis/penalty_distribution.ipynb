```python
# club

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\co279\\penalty.csv")
df = df.iloc[:, 3:]

# data filtering ('last'= 1)
df_last_1 = df[df['last'] == 1]

# data filtering ('last'= 0)
df_last_0 = df[df['last'] == 0]

# spliting to score_o, score_x
score_o_1 = df_last_1[df_last_1['score'] == 'o']
score_x_1 = df_last_1[df_last_1['score'] == 'x']

score_o_0 = df_last_0[df_last_0['score'] == 'o']
score_x_0 = df_last_0[df_last_0['score'] == 'x']

# preparing labels
labels = ['score_o', 'score_x']

sizes_1 = [len(score_o_1), len(score_x_1)]
sizes_0 = [len(score_o_0), len(score_x_0)]

colors = ['lightskyblue', 'lightcoral']

# pie chart
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# 'last' = 0
axes[0].pie(sizes_0, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, counterclock=False)
axes[0].set_title('Proportion of last=0')

# 'last' = 1
axes[1].pie(sizes_1, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, counterclock=False)
axes[1].set_title('Proportion of last=1')


plt.show()
```
