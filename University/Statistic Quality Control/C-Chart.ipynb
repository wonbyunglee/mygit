```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
df = pd.read_csv("21_delay.csv")
df = df[['week', 'total']]
df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["total"], linestyle='-', color='blue', label='Delays')
plt.axhline(y=df["total"].mean(), color='red', linestyle='--', label='Mean')
center_line = df["total"].mean()
UCL = center_line + 3 * np.sqrt(center_line)
LCL = max(center_line - 3 * np.sqrt(center_line), 0)

plt.axhline(y=UCL, color='green', linestyle='--', label='UCL')
plt.axhline(y=LCL, color='green', linestyle='--', label='LCL')

plt.xlabel('week')
plt.ylabel('Delays')
plt.title('C-Chart for delays (phase1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['Mean'] = df['total'].mean()
df['UCL'] = df['Mean'] + 3 * np.sqrt(df['Mean'])
new_df = df[df['total'] <= df['UCL']]
new_df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(new_df["total"], linestyle='-', color='blue', label='Delays')
plt.axhline(y=new_df["total"].mean(), color='red', linestyle='--', label='Mean')
center_line = new_df["total"].mean()
UCL = center_line + 3 * np.sqrt(center_line)
LCL = max(center_line - 3 * np.sqrt(center_line), 0)

plt.axhline(y=UCL, color='green', linestyle='--', label='UCL')
plt.axhline(y=LCL, color='green', linestyle='--', label='LCL')

plt.xlabel('week')
plt.ylabel('Delays')
plt.title('new C-Chart for delays (phase1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['total'], bins=range(df['total'].min(), df['total'].max() + 1), align='left', color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram(phase1)')
plt.xlabel('Total Delays')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 포아송 분포의 이론적인 확률 질량 함수 플로팅
lambda_param = df['total'].mean()
x = range(df['total'].min(), df['total'].max() + 1)
pmf_values = poisson.pmf(x, lambda_param)
plt.plot(x, pmf_values * len(df), linestyle='-', color='red', label='Poisson PMF')

plt.legend()
plt.show()
```
