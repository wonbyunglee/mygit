```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df18 = pd.read_csv("18all.csv", encoding='euc-kr')
df19 = pd.read_csv("19all.csv", encoding='euc-kr')
df20 = pd.read_csv("20all.csv", encoding='euc-kr')
df21 = pd.read_csv("21all.csv", encoding='euc-kr')
df22 = pd.read_csv("22all.csv", encoding='euc-kr')
```

```python
df18 = df18[['일자', '총 피해액(백만원)']]
df19 = df19[['일자', '총 피해액(백만원)']]
df20 = df20[['일자', '총 피해액(백만원)']]
df21 = df21[['일자', '총 피해액(백만원)']]
df22 = df22[['일자', '총 피해액(백만원)']]
```

```python
df18['일자'] = pd.to_datetime(df18['일자'])
monthly_mean = df18.groupby(df18['일자'].dt.to_period("M"))['총 피해액(백만원)'].mean()

# 새로운 열로 추가
df18['월별 평균 피해액'] = df18['일자'].dt.to_period("M").map(monthly_mean)
monthly_mean_unique = monthly_mean[~monthly_mean.index.duplicated(keep='first')]

# '일자' 열을 데이터프레임의 인덱스로 설정
monthly_mean_unique = monthly_mean_unique.reset_index()
df18 = monthly_mean_unique
df18
```

```python
dfs = [df18, df19, df20, df21]
phase1 = pd.concat(dfs, ignore_index=True)
phase1
```

```python
phase2 = df22
phase2
```

```python
phase1['일자'] = pd.to_datetime(phase1['일자'].astype(str))

plt.figure(figsize=(10, 6))
plt.plot(phase1['일자'], phase1['총 피해액(백만원)'], linestyle='-')
plt.axhline(phase1['총 피해액(백만원)'].mean(), color='red', linestyle='dashed', label='mean')
XCL = phase1['총 피해액(백만원)'].mean() + 3 * phase1['총 피해액(백만원)'].std()
UCL = phase1['총 피해액(백만원)'].mean() + 2 * phase1['총 피해액(백만원)'].std()
LCL = phase1['총 피해액(백만원)'].mean() - 2 * phase1['총 피해액(백만원)'].std()
plt.axhline(XCL, color='green', linestyle='dashed', label='3sigma')
plt.axhline(UCL, color='orange', linestyle='dashed', label='2sigma')
plt.axhline(LCL, color='orange', linestyle='dashed')

plt.title('X-Chart for Total Damages(phase1)')
plt.xlabel('Date')
plt.ylabel('Total Damages (Million Won)')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
from scipy import stats

plt.figure(figsize=(8, 6))
sm.qqplot(phase1['총 피해액(백만원)'], line='s')
plt.title('QQ-Plot for Total Damages(phase1)')
plt.show()
```

```python
phase1['총 피해액(백만원)'] = np.log(phase1['총 피해액(백만원)'])
phase1['총 피해액(백만원)'] = phase1['총 피해액(백만원)'].replace(-np.inf, 0.001)
phase1
```

```python
dates = phase1['일자']
total_damages = phase1['총 피해액(백만원)']

# R 값을 계산하여 리스트에 저장
R_values = [abs(total_damages[i] - total_damages[i - 1]) for i in range(1, len(total_damages))]

# R 평균과 3시그마 경계 계산
RCL = np.mean(R_values)
R_UCL = RCL + 3 * np.std(R_values)

# R 관리도 그리기
plt.figure(figsize=(10, 4))
plt.plot(dates[1:], R_values, linestyle='-', color='purple', label='R value')
plt.axhline(RCL, color='red', linestyle='dashed', label='mean')
plt.axhline(R_UCL, color='green', linestyle='dashed', label='3sigma')

plt.title('R-Chart for Total Damages (phase1)')
plt.xlabel('date')
plt.ylabel('R value')
plt.legend()
plt.show()
```
